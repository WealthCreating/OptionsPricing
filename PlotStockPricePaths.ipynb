The BS SDE is given by the following:
\begin{equation}
dS_t = S_t\left(rdt + \sigma dW_t\right)
\end{equation}
where $W_t$ is a regular Brownian motion.

```python
testPlot.plotHestonPaths(kappa = 0.75, eta = 0.0516, theta = 0.0238, rho = -0.9, vol0 = 0.2)
```

The Heston model is defined by two processes, i.e. a stock price process and a volatility process
(constant volatility assumption from BS model is relaxed).
The stock price process under Heston assumptions is given here below:
\begin{equation}\label{Heston 1}
dS_t = S_t\left(rdt + \sqrt{\sigma_t^2} dW_t\right)
\end{equation}

```python
testPlot.plotHestonVolatilityPaths(kappa = 0.75, eta = 0.0516, theta = 0.0238, rho = -0.9, vol0 = 0.2)
```

The volatility process of Heston model is:
\begin{equation}\label{Heston 2}
d\sigma_t^2 = \kappa\left(\eta-\sigma_t^2\right)dt+\theta\sqrt{\sigma_t^2}d\tilde{W}_t
\end{equation}
where:

\begin{enumerate}
\item \kappa: speed of mean reversion of volatility, i.e. the speed at which the process reverses back to its mean
\item \eta: level of mean reversion of volatility
\item \theta: vol-of vol
\item \tilde{W}_t: Brownian motion that is correlated with the Browniann motion present in the stock price process of Heston model.
    One uses Cholesky decomposition on a given covariance matrix to generate Brownian motions with the desired correlation.
\end{enumerate}

The Merton Jump diffusin SDE is an extension of the BS SDE with an extra component:
\begin{equation}
dS_t = S_t\left(rdt + \sigma dW_t\right + J_tdN_t)  
\end{equation}
where:
\begin{itemize}
\item $W_t$ is a regular Brownian motion.
\item $J_t$ determines the size of the jump at time t and $log(1+J_t)$ follows a normal distribution $\sim N\left(log(1+\mu)-\frac{\delta^2}{2},\delta^2\right)$.
\item $N_t$ is a Poisson process with intensity measure $\lambda t$
\end{itemize}
Please refer to Hilpisch(2014) for more detail.
